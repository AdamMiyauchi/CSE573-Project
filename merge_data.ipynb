```python
def historical_sentiment(df):
    for i in range(1,26):
        df[f'entity_pos_sent_minus_{i}'] = df['entity_pos_sent'].shift(i)
        df[f'entity_neg_sent_minus_{i}'] = df['entity_neg_sent'].shift(i)
        df[f'entity_net_sent_minus_{i}'] = df['entity_net_sent'].shift(i)
    return df
```

Apple: aggregated sentiment, historical price, price direction, percent change

```python
# Mean news sentiment aggregation
aapl_news_mean = (df_news[(df_news['title_aapl'] == True) | (df_news['text_aapl'] == True)]
    .groupby('prediction_date')
    .apply(lambda df: df[['entity_pos_sent', 'entity_neg_sent', 'entity_net_sent']].mean())
    .reset_index()
    .sort_values('prediction_date')
)
aapl_news_mean = historical_sentiment(aapl_news_mean)

(aapl_news_mean.merge(df_apple, how='inner', left_on='prediction_date', right_on='date')
    .drop(['prediction_date', 'time', 'open', 'high', 'low', 'close', 'volume'], axis=1)
    .sort_values('date', ascending=True)
    .to_csv('data/preprocessed/Aapl_SentimentAggMeanPriceHistDirection.csv', index=False)
)
print('Apple Mean Sentiment')
pd.read_csv('data/preprocessed/Aapl_SentimentAggMeanPriceHistDirection.csv').head()

# Sum news sentiment aggregation
aapl_news_sum = (df_news[(df_news['title_aapl'] == True) | (df_news['text_aapl'] == True)]
    .groupby('prediction_date')
    .apply(lambda df: df[['entity_pos_sent', 'entity_neg_sent', 'entity_net_sent']].sum())
    .reset_index()
    .sort_values('prediction_date')
)
aapl_news_sum = historical_sentiment(aapl_news_sum)

(aapl_news_sum.merge(df_apple, how='inner', left_on='prediction_date', right_on='date')
    .drop(['prediction_date', 'time', 'open', 'high', 'low', 'close', 'volume'], axis=1)
    .sort_values('date', ascending=True)
    .to_csv('data/preprocessed/Aapl_SentimentAggSumPriceHistDirection.csv', index=False)
)
print('Apple Sum Sentiment')
pd.read_csv('data/preprocessed/Aapl_SentimentAggSumPriceHistDirection.csv').head()
```

Amazon: aggregated sentiment, historical price, price direction, percent change

```python
# Mean news sentiment aggregation
amzn_news_mean = (df_news[(df_news['title_amzn'] == True) | (df_news['text_amzn'] == True)]
    .groupby('prediction_date')
    .apply(lambda df: df[['entity_pos_sent', 'entity_neg_sent', 'entity_net_sent']].mean())
    .reset_index()
    .sort_values('prediction_date')
)
amzn_news_mean = historical_sentiment(amzn_news_mean)

(amzn_news_mean.merge(df_amazon, how='inner', left_on='prediction_date', right_on='date')
    .drop(['prediction_date', 'time', 'open', 'high', 'low', 'close', 'volume'], axis=1)
    .sort_values('date', ascending=True)
    .to_csv('data/preprocessed/Amzn_SentimentAggMeanPriceHistDirection.csv', index=False)
)
print('Amazon Mean Sentiment')
pd.read_csv('data/preprocessed/Amzn_SentimentAggMeanPriceHistDirection.csv').head()

# Sum news sentiment aggregation
amzn_news_sum = (df_news[(df_news['title_amzn'] == True) | (df_news['text_amzn'] == True)]
    .groupby('prediction_date')
    .apply(lambda df: df[['entity_pos_sent', 'entity_neg_sent', 'entity_net_sent']].sum())
    .reset_index()
    .sort_values('prediction_date')
)
amzn_news_sum = historical_sentiment(amzn_news_sum)

(amzn_news_sum.merge(df_amazon, how='inner', left_on='prediction_date', right_on='date')
    .drop(['prediction_date', 'time', 'open', 'high', 'low', 'close', 'volume'], axis=1)
    .sort_values('date', ascending=True)
    .to_csv('data/preprocessed/Amzn_SentimentAggSumPriceHistDirection.csv', index=False)
)
print('Amazon Sum Sentiment')
pd.read_csv('data/preprocessed/Amzn_SentimentAggSumPriceHistDirection.csv').head()
```

### BERT direction prediction data preprocessing

```python
aapl_bert_feat = (df_news[(df_news['title_aapl'] == True) | (df_news['text_aapl'] == True)][['title', 'prediction_date']]
    .merge(df_apple[['date', 'direction']], how='inner', left_on='prediction_date', right_on='date')
    .drop('prediction_date', axis=1)
)
print('Apple BERT features')
aapl_bert_feat.shape
aapl_bert_feat.head()
aapl_bert_feat.to_csv('data/apple_bert.csv', index=False)

amzn_bert_feat = (df_news[(df_news['title_amzn'] == True) | (df_news['text_amzn'] == True)][['title', 'prediction_date']]
    .merge(df_amazon[['date', 'direction']], how='inner', left_on='prediction_date', right_on='date')
    .drop('prediction_date', axis=1)
)
print('Amazon BERT features')
amzn_bert_feat.shape
amzn_bert_feat.head()
amzn_bert_feat.to_csv('data/amazon_bert.csv', index=False)
```

#### Combine all news titles for each prediction date. Will be used to fine-tune train a BERT model

```python
aapl_titles = (df_news[(df_news['title_aapl'] == True) | (df_news['text_aapl'] == True)].groupby('prediction_date')
    .apply(lambda df: ' '.join(df['title']))
    .reset_index()
    .rename({0: 'combined_titles'}, axis=1)
)

amzn_titles = (df_news[(df_news['title_amzn'] == True) | (df_news['text_amzn'] == True)].groupby('prediction_date')
    .apply(lambda df: ' '.join(df['title']))
    .reset_index()
    .rename({0: 'combined_titles'}, axis=1)
)

aapl_bert_feat = (aapl_titles.merge(df_apple, how='inner', left_on='prediction_date', right_on='date')[['date', 'combined_titles', 'direction']]
    .sort_values('date', ascending=True)
)
aapl_bert_feat.to_csv('data/apple_bert_combined.csv', index=False)

amzn_bert_feat = (amzn_titles.merge(df_amazon, how='inner', left_on='prediction_date', right_on='date')[['date', 'combined_titles', 'direction']]
    .sort_values('date', ascending=True)
)
amzn_bert_feat.to_csv('data/amazon_bert_combined.csv', index=False)
```
